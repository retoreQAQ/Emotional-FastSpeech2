```python
# ÊâπÈáèÂêàÊàê
#!/bin/bash

ckpt="finetune_1200000"
text="I am so hungry that I have to eat something."

# ÂÆö‰πâÊÉÖÁª™Á±ªÂà´ÔºàÂèØÊåâ‰Ω†ÁöÑÊï∞ÊçÆÈõÜÂÆûÈôÖ‰øÆÊîπÔºâ
emotions=("A", "C", "D", "F", "H", "N", "O", "S", "U", "X")  # Angry, Happy, Sad, Neutral

# valence Âíå arousal ËøûÁª≠ÂÄºÔºàËåÉÂõ¥ 1.0‚Äì5.0 / 1.0‚Äì7.0Ôºâ
valences=(1.0 4.0 7.0)
arousals=(1.0 4.0 7.0)

for emo in "${emotions[@]}"; do
  for val in "${valences[@]}"; do
    for aro in "${arousals[@]}"; do
      tag="${emo}_V${val}_A${aro}"
      echo "üîä Synthesizing: $tag"
      python synthesize.py \
          --text $text \
          --restore_step $ckpt \
          --mode single \
          -p config/MSP/preprocess.yaml \
          -m config/MSP/model.yaml \
          -t config/MSP/train_finetune.yaml \
          --emotion $emo \
          --valence $val \
          --arousal $aro \
          --pitch_control 1.0 \
          --energy_control 1.0 \
          --duration_control 1.0
    done
  done
done

echo "‚úÖ All synthesis finished."
```
